```python
from fairchem.core.datasets import create_dataset, data_list_collater
from torch.utils.data import DataLoader

# Create dataset using the config-based API
# Use "ase_db" format which supports LMDB databases via ASE
config = {
    "src": "./data/omol25/train",
    "format": "ase_db"  # ASE database format supports LMDB backends
}

train_dataset = create_dataset(config, split="train")
train_loader = DataLoader(
    train_dataset,
    batch_size=32,
    shuffle=True,
    collate_fn=data_list_collater,
    num_workers=0  # Set to >0 if you want parallel data loading
)
```
